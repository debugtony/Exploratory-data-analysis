```python
# define your hypothesis

null = 'There is no correlation between age and balance'
alternative = 'There is a correlation between age and balance'

# define your alpha level
alpha = 0.05

# decision rule
# if p < alpha: reject null hypothesis else: fail to reject null hypothesis

# calculate the correlation coefficient and get the p-value

test_stats, pvalue = stats.spearmanr(data.age, data.balance)

# conclusion

if pvalue < alpha:
    print(f'{alternative} as the p-value ({pvalue}) < alpha ({alpha}) @ correlation coefficient of {test_stats}')
else:
    print(f'{null} as the p-value ({pvalue}) > alpha ({alpha}) @ correlation coefficient of {test_stats}')
```

```python
px.histogram(data_frame=data, x = 'marital', y = 'balance',
             title = 'Bar chart of Marital vs balance', histfunc = 'avg')
```

```python
# check if marital status has an effect on balance


# define your hypothesis
null = 'There is no difference in deposit across marital status'
alternative = 'There is a difference in deposit across marital status'

# define your alpha level
alpha = 0.05

# decision rule
# if p < alpha: reject null hypothesis else: fail to reject null hypothesis

# calculate the correlation coefficient and get the p-value

married = data[data.marital == 'married'].balance
single = data[data.marital == 'single'].balance
divorced = data[data.marital == 'divorced'].balance

test_stats, pvalue = stats.f_oneway(married, single, divorced)

# conclusion

if pvalue < alpha:
    print(f'{alternative} with pvalue of {pvalue} and conclude that there is a difference in deposit across marital status')
else:
    print(f'{null} and conclude that there is no difference in deposit across marital status')
```

# Hypothesis Testing: ANOVA for Marital Status and Balance

**Null hypothesis:** There is no difference in balance across marital status.
**Alternative hypothesis:** There is a difference in balance across marital status.

**Alpha level:** 0.05.

**Decision rule:**
- If p < alpha: reject null hypothesis.
- Else: fail to reject null hypothesis.

Conducting one-way ANOVA:

Test statistics: {{test_stats}}
p-value: {{pvalue}}

**Conclusion:**
{% if pvalue < alpha %}
With a p-value of {{pvalue}}, we reject the null hypothesis. Thus, there is a statistically significant difference in balance across marital status categories. Further analysis, such as post-hoc tests, may be conducted to determine which groups differ significantly.
{% else %}
The p-value of {{pvalue}} is greater than the alpha level. Hence, we fail to reject the null hypothesis. This suggests that there is no significant difference in balance across marital status categories.
{% endif %}

Please note that these results are based on the assumption of the ANOVA model being met and may require further investigation or validation.

#### Chisquare test

##### check if there is an association between marital status and loan

```python
px.histogram(data_frame = data, x = 'marital', color = 'loan',
             barmode = 'group', title = 'Bar chart of loan and marital status')
```

```python
# set hypothesis

null = 'There is no association between marital status and loan'
alternative = 'There is an association between marital status and loan'

# choose alpha
alpha = 0.05

marital_loan = pd.crosstab(data['marital'], data['loan'])

test, pvalue, dof, exp = stats.chi2_contingency(marital_loan)

# conclusion

if pvalue < alpha:
    print(f'{alternative} @ pvalue = {pvalue}')
else:
    print(f'{null} @ pvalue = {pvalue}')
```

# Hypothesis Testing: Chi-square Test for Marital Status and Loan Association

**Null hypothesis:** There is no association between marital status and loan.
**Alternative hypothesis:** There is an association between marital status and loan.

**Alpha level:** 0.05.

Conducting Chi-square test:

Test statistics: Chi-square value = {{test}}, p-value = {{pvalue}}

**Conclusion:**
{% if pvalue < alpha %}
With a p-value of {{pvalue}}, we reject the null hypothesis. Hence, there is evidence to suggest an association between marital status and loan.
{% else %}
The p-value of {{pvalue}} is greater than the chosen alpha level. Thus, we fail to reject the null hypothesis. Therefore, we do not find sufficient evidence to conclude that there is an association between marital status and loan.
{% endif %}

Chi-square test is used to examine whether there is a significant association between two categorical variables. In this case, the analysis suggests that marital status might be associated with loan status. However, further investigation or analysis may be necessary to understand the nature and strength of this association.
